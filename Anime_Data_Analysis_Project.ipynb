### Program Code

```python
import numpy as np
import pandas as pd
import plotly.graph_objects as go
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from tqdm import tqdm
import plotly.graph_objs as go
import plotly.io as pio
pio.renderers.default = "notebook_connected"
```

#### _Data Import_

```python
anime_df = pd.read_csv('anime-dataset-2023.csv')
```

#### Feature Engineering

The **add_total** function is used to add a new column **Total(dataframe:int)** to the dataframe which is the sum of the column **Type(datatype:string)**

```python
def add_total(anime_types):
  anime_types['Total'] = anime_types.sum(axis=1)
  return anime_types
```

The **Duration(datatype:string)** is converted to **Duration(datatype:float)** using **str_to_float** function

```python
def str_to_float(anime_df):
  anime_df['Duration'] = anime_df['Duration'].astype(str).str.extract(r'(\d+)').astype(float)
  return anime_df
```

#### Data Cleansing

```python
anime_df.isnull().sum()
anime_df = anime_df.drop(['Other name', 'Image URL', 'Licensors'], axis=1)

anime_df[['Start_Year', 'End_Year']] = anime_df['Aired'].str.split(' to ', expand=True)
anime_df['Start_Year'] = anime_df['Start_Year'].str.extract(r'(\d{4})', expand=False)
anime_df['Start_Year'] = pd.to_numeric(anime_df['Start_Year'], errors='coerce')
anime_df = anime_df.dropna(subset=['Start_Year'])
```

#### Percentage of Missing Values

```python
def anime_table(req_df, cols_list):
    fig = go.Figure(data=[go.Table(header=dict(values=list(req_df.columns),
                  fill_color='paleturquoise', align='center'),
                  cells=dict(values=cols_list, fill_color='lavender',
                  align='center'), columnwidth=50)])
    fig.update_layout(height=180, margin=dict(r=5, l=5, t=5, b=5))
    return fig.show()
```

```python
missing_value_df = pd.DataFrame({'Column name': anime_df.columns,
                                'Missing record count': anime_df.eq('UNKNOWN').sum(),
                                'Missing record percentage(%)': round(anime_df.eq('UNKNOWN').sum()
                                * 100 / len(anime_df),
                                2)}).sort_values(by='Missing record percentage(%)'
        , ascending=False)

anime_table(missing_value_df, [missing_value_df['Column name'],
             missing_value_df['Missing record count'],
             missing_value_df['Missing record percentage(%)']])
```

We can see that, Premiered column has **77.05%**, English name column has **57.51%**, Producers column  has **52.23%**, Studios column has **40.52%**, Scored By & Score column has **34.62%**, Genres column has **18.87%**, Rank column has **18.26%**, Rating column has **2.2%**, Episodes column has **1.72%**, Type column has **0.05%** of missing values

#### Objective 1: What patterns might be noticed in animation studios across various anime series, within a specific anime subgenre, and among a collection of popular fan-favorite shows?

```python
def obj1(anime_df):
  anime_df = anime_df[anime_df['Studios'] != "UNKNOWN"]
  anime_df = anime_df[anime_df['Type'] != "UNKNOWN"]
  anime_series = anime_df['Studios'].value_counts().head(15)
  plot1 = go.Figure(data=[go.Bar(x=anime_series.index, y=anime_series)])
  plot1.update_layout(
  title='Studios based on Animes',
  xaxis_title='Studio',
  yaxis_title='Animes'
  )
  plot1.show()
  ###########################################################################################
  fan_favorites = anime_df.sort_values('Favorites', ascending=False)
  top_fan_favorites = fan_favorites.groupby('Studios')['Favorites'].sum().nlargest(15)
  plot2 = go.Figure(data=[go.Bar(x=top_fan_favorites.index, y=top_fan_favorites)])
  plot2.update_layout(
  title='Studios Based on Favorites',
  xaxis_title='Studio',
  yaxis_title='Fans'
  )
  plot2.show()
  ##########################################################################################
  anime_types = anime_df.groupby('Studios')['Type'].value_counts().unstack().fillna(0)
  anime_types = add_total(anime_types)
  top_anime_types = anime_types.sort_values('Total', ascending=False).head(15)
  plot3 = go.Figure(data=[
  go.Bar(name=anime_type, x=top_anime_types.index, y=top_anime_types[anime_type])
  for anime_type in top_anime_types.columns[:-1]
  ])
  plot3.update_layout(
  title='Studios Based on Anime Types',
  xaxis_title='Studio',
  yaxis_title='Animes'
  )
  plot3.show()
```

#### Objective 2: Analyze the average duration of episodes and the impact that has on popularity?

```python
def obj2(anime_df):
  anime_df = str_to_float(anime_df);
  average_duration_epi_vs_pop = anime_df.groupby('Duration')['Popularity'].mean().reset_index()
  plot = go.Figure(data=go.Scatter(
  x=average_duration_epi_vs_pop['Duration'],
  y=average_duration_epi_vs_pop['Popularity'],
  mode='markers',
  marker=dict(
  size=average_duration_epi_vs_pop['Popularity'] / 9000,
  sizemode='diameter',
  sizeref=0.1,
  sizemin=1,
  color=average_duration_epi_vs_pop['Duration'],
  colorscale='Viridis',
  showscale=True
  ),
  text=average_duration_epi_vs_pop['Duration'],
  hovertemplate='Duration: %{text} min<br>Popularity: %{y:.2f}<extra></extra>'
  ))
  plot.update_layout(
  title='Average Duration of Episodes vs Popularity',
  xaxis_title='Duration (minutes)',
  yaxis_title='Popularity',
  )
  plot.show()
```

#### Objective 3: Identify the top anime studios based on favorite ratio of their produced anime?

Formula:

**Favorite Ratio=
Number of Favorites/
Number of Members**

Number of Favorites: The total number of times users marked the anime as a favorite.<br>
Number of Members: The total number of members who have added the anime to their list on the platform. 


```python
def plot_top_studios_by_favorite_ratio(anime_df):
    tqdm.pandas()
    filtered_df = anime_df.dropna(subset=['Studios', 'Favorites'])
    filtered_df['Favorite_Ratio'] = filtered_df['Favorites'] / filtered_df['Members']
    filtered_df['Favorite_Ratio'] = filtered_df['Favorite_Ratio'].replace([np.inf, -np.inf, np.nan], 0)
    studio_favorite_ratio = filtered_df.groupby('Studios')['Favorite_Ratio'].mean().reset_index()
    sorted_studios = studio_favorite_ratio.sort_values(by='Favorite_Ratio', ascending=True).tail(10)
    plt.figure(figsize=(14, 7))
    bar_colors = plt.cm.viridis(sorted_studios['Favorite_Ratio'] / sorted_studios['Favorite_Ratio'].max())
    bars = plt.barh(sorted_studios['Studios'], sorted_studios['Favorite_Ratio'], color=bar_colors)
    plt.title('Top 10 Anime Studios by Favorite Ratio')
    plt.xlabel('Average Favorite Ratio')
    plt.ylabel('Studios')
    plt.show()
```

#### Objective 4: Identify the popularity trend of anime over time, grouped by source?

```python
def visualize_popularity_trend(anime_df, top_sources_count=3, recent_years_count=5):
    def extract_start_date(aired_str):
        try:
            start_date_str = aired_str.split(' to ')[0]
            return pd.to_datetime(start_date_str)
        except Exception as e:
            return np.nan


    anime_df_filtered = anime_df[anime_df['Source'] != 'Unknown'].copy()


    anime_df_filtered.loc[:, 'Start_Date'] = anime_df_filtered['Aired'].apply(extract_start_date)

    popularity_trend = anime_df_filtered.groupby([anime_df_filtered['Start_Date'].dt.year, 'Source'])['Popularity'].mean().reset_index()
    popularity_trend_sorted = popularity_trend.sort_values(by=['Start_Date', 'Popularity'], ascending=[True, True])

    sns.set(style='whitegrid')
    plt.figure(figsize=(14, 8))
    for source in popularity_trend_sorted['Source'].unique():
        trend_data = popularity_trend_sorted[popularity_trend_sorted['Source'] == source]
        plt.plot(trend_data['Start_Date'], trend_data['Popularity'], label=source)

    plt.yscale('log')
    plt.xlabel('Year')
    plt.ylabel('Average Popularity (Log Scale)')
    plt.title('Popularity Trend Over Time by Source')
    plt.legend()
    plt.show()

    top_sources = anime_df_filtered['Source'].value_counts().head(top_sources_count).index.tolist()

    recent_years_cleaned = anime_df_filtered[
        (anime_df_filtered['Start_Date'].dt.year >= anime_df_filtered['Start_Date'].dt.year.max() - recent_years_count)
    ]

    recent_popularity_cleaned = recent_years_cleaned.groupby('Source')['Popularity'].mean().sort_values(ascending=False)
    recent_popularity_top_sources_cleaned = recent_popularity_cleaned[
        recent_popularity_cleaned.index.isin(top_sources)
    ].reset_index()

    print(f'Detailed key observations for the top {top_sources_count} sources over the last {recent_years_count} years:')
    print(recent_popularity_top_sources_cleaned)
```

#### Objective 5: How do anime ratings and members counts distribute over time, and which rating category dominates user engagement?

```python
def obj5(anime_df):
    anime_df = anime_df[anime_df['Rating'] != "UNKNOWN"]
    ratings_by_year = anime_df.groupby('Start_Year')['Members'].mean().reset_index()

    plt.figure(figsize=(12, 6))
    plt.plot(ratings_by_year['Start_Year'], ratings_by_year['Members'], marker='o', linestyle='-', color='brown')
    plt.title('Distribution of users Over the Years')
    plt.xlabel('Year')
    plt.ylabel('Average Members')
    plt.grid(True)
    plt.show()

    plt.figure(figsize=(12, 12))

    plt.subplot(2, 1, 1)
    for rating_category, group in anime_df.groupby('Rating'):
        count_by_year = group.groupby('Start_Year').size()
        plt.plot(count_by_year.index, count_by_year.values, marker='o', linestyle='-', label=rating_category)

        plt.title('Distribution of Ratings Over the Years')
        plt.xlabel('Year')
        plt.ylabel('Number of Anime')
        plt.legend(title='Rating')
        plt.grid(True)

    plt.subplot(2, 1, 2)
    top_rating = anime_df.sort_values(by='Members', ascending=False)
    plt.bar(top_rating['Rating'].astype(str), top_rating['Members'], color='green')
    plt.xticks(rotation=45, ha='right')
    plt.title('Top Rating by Number of Users')
    plt.ylabel('Number of Users')
    plt.xlabel('Rating Types')
    plt.tight_layout()
    plt.show()
```

#### Objective 6: Determining the most popular anime adapted from specific source materials (eg. manga, light novels).

```python
def visualize_average_popularity_by_source(anime_df):
    anime_df_filtered = anime_df[anime_df['Source'] != 'Unknown']

    average_popularity_by_source = anime_df_filtered.groupby('Source')['Popularity'].mean().sort_values(ascending=True)

    plt.figure(figsize=(10, 6))
    average_popularity_by_source.plot(kind='barh', xlabel='Source Material', ylabel='Average Popularity', title='Impact of Source Material on Popularity')
    plt.show()
```

#### Objective 7: Identify the most popular anime based on factors like popularity and number of scored reviews.

```python
def obj7(anime_df):
  anime_df['Popularity'] = np.where(anime_df['Popularity'] == 'UNKNOWN', np.nan, anime_df['Popularity'])
  anime_df['Scored By'] = np.where(anime_df['Scored By'] == 'UNKNOWN', np.nan, anime_df['Scored By'])
  anime_df['Popularity'] = pd.to_numeric(anime_df['Popularity'], errors='coerce')
  anime_df['Scored By'] = pd.to_numeric(anime_df['Scored By'], errors='coerce')
  anime_df.dropna(subset=['Popularity', 'Scored By', 'Score'], inplace=True)
  features = ['Popularity', 'Scored By']
  target = 'Score'
  anime_df.dropna(subset=features + [target], inplace=True)
  X_train, X_test, y_train, y_test = train_test_split(anime_df[features], anime_df[target], test_size=0.2, random_state=42)
  model = RandomForestRegressor(n_estimators=100, random_state=42)
  model.fit(X_train, y_train)
  y_pred = model.predict(X_test)
  rmse = mean_squared_error(y_test, y_pred, squared=False)
  print('Root Mean Squared Error:', rmse)
  plt.scatter(X_test['Popularity'], X_test['Scored By'], c=y_pred, cmap='viridis')
  plt.colorbar(label='Predicted Score')
  plt.xlabel('Popularity')
  plt.ylabel('Scored By')
  plt.title('Most Popular Anime by Predicted Score')
  plt.show()
```
